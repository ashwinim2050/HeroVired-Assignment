# Task
Analyze the relationships between numerical features in the `car_prices` dataset by calculating and visualizing their correlation matrix, and then summarizing the key correlations observed.

## Identify Numerical Features

### Subtask:
Select only the numerical columns from the `car_prices` DataFrame.


**Reasoning**:
To select only the numerical columns, I will use the `select_dtypes` method on the `car_prices` DataFrame, specifying `include='number'` to get both integer and float types. The result will be stored in a new DataFrame called `numerical_features_df`.



```python
numerical_features_df = car_prices.select_dtypes(include=['number'])
print("Numerical features DataFrame head:")
display(numerical_features_df.head())
```

**Reasoning**:
Now that the numerical features have been identified, the next step is to calculate the correlation matrix for these features to understand the linear relationships between them.



```python
correlation_matrix = numerical_features_df.corr()
print("Correlation Matrix:")
display(correlation_matrix)
```

**Reasoning**:
To better understand the relationships between the numerical features, I will visualize the correlation matrix as a heatmap. This graphical representation will make it easier to identify strong positive and negative correlations.



```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

# Task
Calculate the average selling price of cars by year from the `car_prices` DataFrame.

## Calculate Average Selling Price by Year

### Subtask:
Group the `car_prices` DataFrame by the 'year' column and calculate the average 'sellingprice' for each year.


**Reasoning**:
To calculate the average selling price per year, I will group the `car_prices` DataFrame by the 'year' column and then compute the mean of the 'sellingprice' for each group.



```python
average_price_by_year = car_prices.groupby('year')['sellingprice'].mean()
print("Average selling price of cars by year:\n", average_price_by_year)
```

## Visualize Average Price by Year

### Subtask:
Create a bar plot to visualize the average selling price of cars over the years. The plot will include appropriate labels for the x and y axes, a title, and a clear representation of the trend.


**Reasoning**:
To visualize the average selling price of cars over the years, I will create a bar plot using the `average_price_by_year` Series. This plot will clearly show the trend of car prices across different years with appropriate labels and title.



```python
plt.figure(figsize=(15, 7))
sns.barplot(x=average_price_by_year.index, y=average_price_by_year.values, palette='viridis')
plt.xlabel('Year')
plt.ylabel('Average Selling Price')
plt.title('Average Car Selling Price by Year')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
