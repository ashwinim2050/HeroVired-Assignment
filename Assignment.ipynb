```python
missing_values_count = car_prices.isnull().sum()
print("Missing values count:\n", missing_values_count)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter out columns with no missing values
missing_values_plot = missing_values_count[missing_values_count > 0]

plt.figure(figsize=(12, 6))
sns.barplot(x=missing_values_plot.index, y=missing_values_plot.values)
plt.title('Count of Missing Values per Column')
plt.xlabel('Column Name')
plt.ylabel('Number of Missing Values')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
duplicate_rows_count = car_prices.duplicated().sum()
print(f"Number of duplicate : {duplicate_rows_count}")
```

```python
car_prices_cleaned = car_prices.drop_duplicates()
print(f"Original shape: {car_prices.shape}")
print(f"Shape after dropping duplicates: {car_prices_cleaned.shape}")
duplicate_rows_count_after_removal = car_prices_cleaned.duplicated().sum()
print(f"Number of duplicate rows after removal: {duplicate_rows_count_after_removal}")
```

```python
average_car_price = car_prices['sellingprice'].mean()
print(f"The average car price is: {average_car_price:}")
```

```python
minimum_car_price = car_prices['sellingprice'].min()
print(f"The minimum car price is: {minimum_car_price:}")
```

```python
maximum_car_price = car_prices['sellingprice'].max()
print(f"The maximum car price is: {maximum_car_price:}")
```

```python
unique_car_colors = car_prices['color'].unique()
print(f"Unique car colors: {unique_car_colors:}")
```

```python
unique_car_brands_count = car_prices['make'].nunique()
unique_car_models_count = car_prices['model'].nunique()

print(f"Number of unique car brands: {unique_car_brands_count}")
print(f"Number of unique car models: {unique_car_models_count}")
```

```python
high_priced_cars = car_prices[car_prices['sellingprice'] > 165000]
display(high_priced_cars)
```

```python
top_car_models = car_prices['model'].value_counts().head(5)
print("Top 5 frequently sold car models:\n", top_car_models)
```

```python
average_price_by_make = car_prices.groupby('make')['sellingprice'].mean()
print("Average selling price of cars by brand (make):\n", average_price_by_make)
```

```python
Min_price_by_interior = car_prices.groupby('interior')['sellingprice'].min()
print(f"Minimum selling price of cars by:\n, {Min_price_by_interior}")
```

```python
highest_odometer_cars = car_prices.groupby('year')['odometer'].max().sort_values(ascending=False)
print(f"Highest odometer reading cars by year:\n, {highest_odometer_cars}")
```

```python
car_prices['car_age'] = 2025 - car_prices['year']
car_prices
```

```python
cars_with_condition = car_prices[(car_prices['condition'] >= 48) & (car_prices['odometer']>90000)]
num_cars_with_condition = len(cars_with_condition)
print(f"Number of cars with condition: {num_cars_with_condition}")
```

```python
newer_cars = car_prices[car_prices['year'] > 2013]
average_price_by_state_newer_cars = newer_cars.groupby('state')['sellingprice'].mean()
```

```python
average_price_by_state_newer_cars = newer_cars.groupby('state')['sellingprice'].mean().sort_values(ascending=False).head(1)
print("state with highest car prices for new cars", average_price_by_state_newer_cars)
```

```python
excellent_condition_threshold = car_prices['condition'].quantile(0.80)
excellent_condition_cars = car_prices[car_prices['condition'] >= excellent_condition_threshold]
avg_price_excellent_condition = excellent_condition_cars.groupby('make')['sellingprice'].mean()
best_value_makes = avg_price_excellent_condition.sort_values()
print(f"Top best value cars by make: {best_value_makes}")
```

```python
numerical_features_df = car_prices.select_dtypes(include=['number'])
print("Numerical features DataFrame head:")
display(numerical_features_df.head())
```

```python
correlation_matrix = numerical_features_df.corr()
print("Correlation Matrix:")
display(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
average_price_by_odometer = car_prices.groupby('odometer')['sellingprice'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='odometer', y='sellingprice', data=average_price_by_odometer)
plt.title('Average Selling Price by Odometer Reading')
plt.xlabel('Odometer Reading')
plt.ylabel('Average Selling Price')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
average_price_by_year = car_prices.groupby('year')['sellingprice'].mean()
print("Average selling price of cars by year:\n", average_price_by_year)
```

```python
plt.figure(figsize=(15, 7))
sns.barplot(x=average_price_by_year.index, y=average_price_by_year.values, palette='viridis')
plt.xlabel('Year')
plt.ylabel('Average Selling Price')
plt.title('Average Car Selling Price by Year')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
cars_sold_by_state = car_prices['state'].value_counts().reset_index()
cars_sold_by_state.columns = ['state', 'num_cars_sold']
print("Number of cars sold per state:\n", cars_sold_by_state.head())
```

```python
plt.figure(figsize=(15, 7))
sns.barplot(x='state', y='num_cars_sold', data=cars_sold_by_state, palette='viridis')
plt.xlabel('State')
plt.ylabel('Number of Cars Sold')
plt.title('Number of Cars Sold by State')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

From the graph the `top_3_states` with the highest number of car sales are: **CA**, **FL**, and **TX**.
