```python
missing_values_count = car_prices.isnull().sum()
print("Missing values count:\n", missing_values_count)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter out columns with no missing values
missing_values_plot = missing_values_count[missing_values_count > 0]

plt.figure(figsize=(12, 6))
sns.barplot(x=missing_values_plot.index, y=missing_values_plot.values)
plt.title('Count of Missing Values per Column')
plt.xlabel('Column Name')
plt.ylabel('Number of Missing Values')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
duplicate_rows_count = car_prices.duplicated().sum()
print(f"Number of duplicate : {duplicate_rows_count}")
```

```python
car_prices_cleaned = car_prices.drop_duplicates()
print(f"Original shape: {car_prices.shape}")
print(f"Shape after dropping duplicates: {car_prices_cleaned.shape}")
duplicate_rows_count_after_removal = car_prices_cleaned.duplicated().sum()
print(f"Number of duplicate rows after removal: {duplicate_rows_count_after_removal}")
```

```python
average_car_price = car_prices['sellingprice'].mean()
print(f"The average car price is: {average_car_price:}")
```

```python
minimum_car_price = car_prices['sellingprice'].min()
print(f"The minimum car price is: {minimum_car_price:}")
```

```python
maximum_car_price = car_prices['sellingprice'].max()
print(f"The maximum car price is: {maximum_car_price:}")
```

```python
unique_car_colors = car_prices['color'].unique()
print(f"Unique car colors: {unique_car_colors:}")
```

```python
unique_car_brands_count = car_prices['make'].nunique()
unique_car_models_count = car_prices['model'].nunique()

print(f"Number of unique car brands: {unique_car_brands_count}")
print(f"Number of unique car models: {unique_car_models_count}")
```

```python
high_priced_cars = car_prices[car_prices['sellingprice'] > 165000]
display(high_priced_cars)
```

```python
top_car_models = car_prices['model'].value_counts().head(5)
print("Top 5 frequently sold car models:\n", top_car_models)
```

```python
average_price_by_make = car_prices.groupby('make')['sellingprice'].mean()
print("Average selling price of cars by brand (make):\n", average_price_by_make)
```

```python
Min_price_by_interior = car_prices.groupby('interior')['sellingprice'].min()
print(f"Minimum selling price of cars by:\n, {Min_price_by_interior}")
```

```python
highest_odometer_cars = car_prices.groupby('year')['odometer'].max().sort_values(ascending=False)
print(f"Highest odometer reading cars by year:\n, {highest_odometer_cars}")
```
